## Jupyter Notebook
### How to Use:

Using Jupyter-hub/notebook can be quite jarring during your first use so we are going to go through some basic commands you will need to get you started.

### Accessing a Notebook:

When trying to open a jupyter notebook file always look for a .ipynb file. This file extension is the extension for an actual jupyter notebook file. So when trying to open one of the labs ignore the other file extensions (unless the tutorial tells you to look at another) and click on the .ipynb file to get started.

### Running a Jupyter Notebook File:

To actually run a notebook file all you have to do is hit the RUN button in the menu bar at the top. This button will run a single cell/block of the file. Alternatively you can use SHIFT + ENTER to run the current cell of the notebook file as well.



Each cell is either a markdown or code cell. Markdown cells contain information and text that is simply just text, running one of these cells just moves to the next cell. Code cells is where the magic happens, they contain code in the language specified (here that will be R or Python) and when you run the cell, you run the code. There is a symbol to the left of each cell labeled IN[]. If the IN[] area has a * in it that means the code is still running, so just wait for it to finish; a lot of this code deals with large amounts of data so some labs may take a few minutes to run. If the IN[] area has a number in it, that means that the code has run and you can move onto the next cell.

### Kernels:

There are two kernels available: R and Python. The labs you will be running will alternate between the two languages and should autoset to the language used in the .ipynb file so don't worry about setting the kernel each time. The language that the current file is using is shown in the top left corner right next to the TRUSTED or NOT TRUSTED button.

When you create a new jupyter notebook, you will need to select which language to use; make sure to pick the right one.

IMPORTANT: To lower the load on our servers, please shut down your server after you are finished working on the labs. To do this, click "Control Panel" in the top right of your screen and then click the red "Shut Down Server" button. This will help keep our servers from overloading when you perform heavy computations.

### Permissions:

There are permissions for your files so you might be able to save and write to certain folders. In your home directory (the directory you start in) you are able to create, run, and save any files. All of the outputs from the labs will be saved to the outputs folder, so if the labs save a picture or dataset somewhere, that is where you will find it. The labs folder is the only folder where you will not be able to save any changes made. If you plan on adding code to expand on the labs, the following steps will guide you to copy the lab to a new location.

### How to save changes to labs
#### Step 1: Download the lab and all other files you would like to copy
Go to the lab you want to copy and click on the checkbox on the left side and then click download. If there is any other files located in the lab folder, make sure to download those as well.

#### Step 2: Create a new folder to place your lab inside
You should do this in your home directory (with the labs and outputs folders). Do this by clicking new in the top right and then selecting folder

Rename it by checking the box to the left of it and selecting rename

#### Step 3: Upload the lab and all other files you downloaded to the new folder
Click upload in the top right and then select the files that you downloaded. They should be located in your downloads folder. 

There may be a warning message if the file is very large, that is okay so just click upload anyways. This will display the file in the folder, then all you have to do is click upload once more on the right side to be able to use the file. Make sure to do this with every file that you need. 
Now you will be able to save all changes made to this document. Note that you may need to change some file paths to access the data. 

So now you have all the basic information to start using jupyternotebook, there are many more tools that can be used so feel free to mess around with them to figure out what they do. Just remember to save your work whenever you make changes to a notebook; its the left most icon in the toolbar at the top with the floppy disk symbol.

#### Step 2: Clone a Repository from Github

1. Find the github repository that you want to clone. Click on the green "clone or download" button in the top right of the page. Copy the link that appears.


2. Open a Terminal in Jupyter Notebook.

3. Create a folder to put the repository in by typing "mkdir <REPO_NAME>" replacing <REPO_NAME> with the desired name. (Note that you could do this outside of the terminal as shown above if you prefer)

4. Type "cd <REPO_NAME>" to enter the folder.

5. Type "git clone <REPO_LINK>" replacing <REPO_LINK> with the link you copied in step 1.

6. Now you will be able to access any tools on github.
